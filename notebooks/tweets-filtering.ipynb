# Filtering tweets for the web app

```python
from twitter_scraping.twitter_scraping import TwitterScraper, TweetsFilter
import daiquiri
import sys
import logging
import os.path as path
import pandas as pd
import datetime

%load_ext autoreload
```

```python
LOGS_DIR = '../logs/'

daiquiri.setup(
    level=logging.INFO,
    outputs=(
        daiquiri.output.Stream(sys.stdout),
        daiquiri.output.File(
            path.join(LOGS_DIR, 'update_tweets.log'),
            formatter=daiquiri.formatter.TEXT_FORMATTER
        )
    )
)

logger = daiquiri.getLogger(__name__)
```

```python
scraper = TwitterScraper(logger)
```

```python
scraper.load_tweets()

scraper.tweets_df.shape
```

```python
%autoreload 2

tweets_filter = TweetsFilter(scraper.tweets_df)
```

Check filtering by companies.

```python
companies_list = ['GamesWorkshop', 'UKGamesWorkshop']

# tweets_filter.filter_by_company(scraper.tweets_df, companies_list)
tweets_filter.filter_tweets(companies_list=companies_list)
```

```python
tweets_filter.filter_by_company(scraper.tweets_df, companies_list).equals(tweets_filter.filter_tweets(companies_list=companies_list))
```

Check filtering by keywords.

```python
keywords_list = ['time', 'picture']

tweets_filter.filter_tweets(keywords_list=keywords_list)
```

```python
tweets_filter.filter_by_keyword(scraper.tweets_df, keywords_list).equals(tweets_filter.filter_tweets(keywords_list=keywords_list))
```

Check filtering by date range.

```python
# dates_range = [pd.to_datetime('2018-10-15'), pd.Timestamp.today()]
dates_range = [pd.to_datetime('2018-10-15'), pd.to_datetime('2018-10-17')]

tweets_filter.filter_tweets(dates_range=dates_range)
```

```python
tweets_filter.filter_by_date(tweets_filter.tweets_df, dates_range).equals(tweets_filter.filter_tweets(dates_range=dates_range))
```

Check combined filtering, reindexing and warning about a company name not being present among the Twitter handles.

```python
companies_list = ['Experian', 'Avast', 'avast']
dates_range = [pd.to_datetime('2018-10-01'), pd.Timestamp.now()]
keywords_list = ['business']

tweets_filter.filter_tweets(companies_list=companies_list, dates_range=dates_range, reindex=True)
```
